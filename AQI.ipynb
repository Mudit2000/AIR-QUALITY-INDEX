# Problem Statement

**Can the data help to monitor the Air Quality Index ?**

The key question we look to answer is from the attributes provided in the dataset, can we predict the Air Quality Index.<br><br>
It is important to understand the behaviour of meteorological parameters in the planetary boundary layer because, atmosphere is the medium in which air pollutants are transported away from the source, which is governed by the meteorological parameters such as atmospheric wind speed, wind direction, humidity and temperature.

Air pollutants are being let out into the atmosphere from a variety of sources, and the concentration of pollutants in the ambient air depends not only on the quantities that are emitted but also the ability of the atmosphere, either to absorb or disperse these pollutants.

Our goal is to eventually provide these insights to correctly predict the Air Quality Index based on these attributes.

# Attribute Description

* T - Average Temperature (°C) <br>
* TM - Maximum temperature (°C) <br>
* Tm - Minimum temperature (°C) <br>
* SLP - Atmospheric pressure at sea level (hPa) <br>
* H - Average relative humidity (%) <br>
* VV - Average visibility (Km) <br>
* V - Average wind speed (Km/h) <br>
* VM - Maximum sustained wind speed (Km/h) <br>
* PM 2.5 - Particulate Matter 2.5 <br>

# Model Building

## Train Test Split

Dropping the feature 'VM' because it has least importance.

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,[0,1,2,3,4,5,6]].values
y = df.iloc[:,-1].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('DT', DecisionTreeRegressor(criterion="mse")))
models.append(('KNN', KNeighborsRegressor(n_neighbors=1)))
models.append(('RR', Ridge()))
models.append(('LR', Lasso()))
models.append(('RFR', RandomForestRegressor()))
models.append(('XGB', xgb.XGBRegressor()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
    kfold = RepeatedKFold(n_splits=10, random_state=1)
    cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='r2')
    results.append(cv_results)
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```

## Hyperparameter Tuning

As per the R2 score, RFR and XGB are the best fitted models on the dataset.<br>
Therefore, we will perform Hyperparameter Tuning on both RFR and XGB to find out the best one.

### Random Forest Regressor

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 800, num = 8)]
max_depth = [int(x) for x in np.linspace(5, 25, num = 5)]
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               }
```

```python
rf_random = RandomizedSearchCV(estimator = RandomForestRegressor(), param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 100, cv = 5, verbose=2, random_state=0, n_jobs = -1)
```

```python
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions=rf_random.predict(x_test)
```

#### Evaluating the model performance

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,mean_squared_error,r2_score
```

```python
print('MAE:', mean_absolute_error(y_test, predictions))
print('MSE:', mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions)))
```

```python
rf_regressor = RandomForestRegressor(n_estimators= 600, max_depth= 20)
```

```python
cross_val= cross_val_score(rf_regressor,x_train,y_train,cv=10,scoring='r2').mean()
cross_val
```

```python
print(r2_score(y_test, predictions))
```

### XGBoost Regressor

```python
learning_rate = ['0.05','0.1', '0.2','0.3','0.5','0.6']
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
random_grid = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'learning_rate': learning_rate
               }
```

```python
xg_random = RandomizedSearchCV(estimator = xgb.XGBRegressor(), param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 100, cv = 5, verbose=2, random_state=0, n_jobs = -1)
```

```python
xg_random.fit(x_train,y_train)
```

```python
xg_random.best_params_
```

```python
xg_random.best_score_
```

```python
predictions=xg_random.predict(x_test)
```

#### Evaluating the model performance

```python
plt.scatter(y_test,predictions)
```

```python
print('MAE:', mean_absolute_error(y_test, predictions))
print('MSE:', mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions)))
```
